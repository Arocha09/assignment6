- Logistic distribution (similar to normal): 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.
- Exponential distribution (similar to log-normal):
$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.
- Negative Binomial (similar to Poisson): The probability mass function for positive integers is:
$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?
- Categorical (similar to Bernoulli): The probability mass function over $k = 1, 2, ..., K$ categories is
$$
f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

Logistic Distribution:

A function has a logistic distribution if its CDF is: 
$$
F(x; \sigma) = \dfrac{1}{1+e^{-x/\sigma}} 
$$
with $\sigma >0$.

The support is $(-\infty, \infty)$, because as $x \rightarrow \infty$, $F(x; \sigma) \rightarrow 1$, and as $x \rightarrow -\infty$, $F(x; \sigma) \rightarrow 0$.

$$
F'(x; \sigma) =
\dfrac{d}{dx} \dfrac{1}{1+e^{-x/\sigma}}
= -1(1+e^-x/\sigma)^{-2}*-e^{-x/\sigma}*\dfrac{1}{\sigma}
= \dfrac{e^{-x/\sigma}}{\sigma(1+e^{-x/\sigma})^2}
$$

```python
grid = np.linspace(-6,6,100)

sigmas = [0.5, 1, 2]
## Plot the density and distribution

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for sigma in sigmas:
    pdf = sp.stats.logistic.pdf(grid, scale=sigma)
    axes[0].plot(grid, pdf, label=f'sigma={sigma}')

axes[0].set_xlabel('x')
axes[0].set_ylabel('Density')
axes[0].set_title('PDF of Logistic Distribution')
axes[0].legend()

plt.figure(figsize=(8,5))
for sigma in sigmas:
    cdf = sp.stats.logistic.cdf(grid, loc=0, scale=sigma)
    axes[1].plot(grid, cdf, label=f'sigma={sigma}')
axes[1].set_title('Logistic CDF for different sigma values')
axes[1].set_xlabel('x')
axes[1].set_ylabel('CDF')
axes[1].legend()
fig.show()
```

```python
sample = sp.stats.logistic.rvs(size=50000,random_state=10)
pdf =  sp.stats.logistic.pdf(grid, scale=1)
fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
sns.lineplot(x=grid,y=pdf,ax=axes[0],label='pdf',linestyle='--')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
sns.lineplot(x=grid,y=cdf,ax=axes[1],label='cdf',linestyle='--')
axes[1].set_title('Sample ECDF')
plt.show()
```

Exponential Distribution:

$$
F_X(x) = \begin{cases}
0, & x<0 \\
1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.

The support of this distribution is $[0, \infty]$ This is because everything larger than zero has a non zero probability.

$$
F_X'(x) = \begin{cases}
0, & x<0 \\
\dfrac{d}{dx}1 - e^{-\lambda x}, & x \ge 0,
\end{cases}
$$
with $\lambda > 0$.

$$
\dfrac{d}{dx}1 - e^{-\lambda x} =
\lambda e^{-\lambda x}
$$

```python
grid = np.linspace(0,12,100)

lambdas = [0.5, 1, 2]
## Plot the density and distribution

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for l in lambdas:
    pdf = sp.stats.expon.pdf(grid, scale=l)
    axes[0].plot(grid, pdf, label=f'lambda={l}')

axes[0].set_xlabel('x')
axes[0].set_ylabel('Density')
axes[0].set_title('PDF of Expon Distribution')
axes[0].legend()

plt.figure(figsize=(8,5))
for l in lambdas:
    cdf = sp.stats.expon.cdf(grid, loc=0, scale=l)
    axes[1].plot(grid, cdf, label=f'lambda={l}')
axes[1].set_title('Expon CDF for different lambda values')
axes[1].set_xlabel('x')
axes[1].set_ylabel('CDF')
axes[1].legend()
fig.show()
```

Negative Binomial Distribution

$$
f(k;r,p) = \dfrac{(k+r-1)!}{k!(r-1)!}(1-p)^k p^r, \quad \text{ for $k=0,...n$}
$$
You can interpret this as follows: Flip a coin that comes up heads with probability $p$ until you get $r$ heads, and then stop. What is the probability of stopping at each $k=0,1,2,...$?

support is $[0, \infty)$ because for each k there is a non zero probability of stopping and k can take any value.

```python
grid = np.linspace(0,12,100)


n, p = 5, 0.5
lb, ub = sp.stats.nbinom.support(n, p)

r_values = [5, 10]        
p_values = [0.3, 0.5]     
k = np.arange(0, 30)
## Plot the density and distribution

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for r, p in zip(r_values, p_values):
    pmf = sp.stats.nbinom.pmf(k, r, p)
    axes[0].plot(k, pmf, marker='o', linestyle='-', label=f'r={r}, p={p}')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Density')
axes[0].set_title('PMF of NBinom Distribution')
axes[0].legend()

plt.figure(figsize=(8,5))
for r, p in zip(r_values, p_values):
    cdf = sp.stats.nbinom.cdf(k, r, p)
    axes[1].step(k, cdf, where='mid', label=f'r={r}, p={p}')
axes[1].set_title('NBinom CDF for different values')
axes[1].set_xlabel('x')
axes[1].set_ylabel('CDF')
axes[1].legend()
fig.show()
```

```python
sample = sp.stats.nbinom.rvs(n=5, p = 0.5, size=50000,random_state=10)

# Side-by-side plot
fig, axes = plt.subplots(1, 2, figsize=(14,5))
r=5
p=0.5
pmf = sp.stats.nbinom.pmf(k, r, p)
# Left: histogram vs PMF
sns.histplot(sample, bins=k, stat='probability', discrete=True, ax=axes[0], color='lightblue', label='Sample histogram')
axes[0].plot(k, pmf, 'r--', label='PMF')
axes[0].set_title('Negative Binomial Sample vs PMF')
axes[0].set_xlabel('Number of failures')
axes[0].set_ylabel('Probability')
axes[0].legend()

cdf= sp.stats.nbinom.cdf(k, r, p)
# Right: ECDF vs CDF
sns.ecdfplot(sample, ax=axes[1], label='Sample ECDF')
axes[1].plot(k, cdf, 'r--', label='CDF')
axes[1].set_title('Negative Binomial Sample')
axes[1].set_xlabel('Number of failures')
axes[1].set_ylabel('Cumulative probability')
axes[1].legend()

plt.tight_layout()
plt.show()
```

f(k;p_1,...,p_K) = p_1^{k=1}p_2^{k=2}...p_K^{k=K}
$$
where $0 \le p_i \le 1$ and $ \sum_{k=1}^K p_k = 1$.

Support is $[1, K]$

```python
K=10

cats = np.random.rand(K)

p = cats/np.sum(cats)
print(p)
print(p.sum())
```

```python

K = 4
categories = np.arange(1, K+1)

# Try multiple probability sets
prob_sets = [
    [0.1, 0.2, 0.3, 0.4],
    [0.25, 0.25, 0.25, 0.25],  # uniform
    [0.5, 0.2, 0.2, 0.1]       # skewed
]

for i, p in enumerate(prob_sets, 1):
    p = np.array(p)
    assert np.isclose(p.sum(), 1.0), "Probabilities must sum to 1"
    
    # Draw samples
    sample = np.random.choice(categories, size=1000, p=p)
    
    # Empirical PMF
    counts = np.bincount(sample, minlength=K+1)[1:]
    empirical_pmf = counts / counts.sum()
    
    # Theoretical and empirical CDF
    theoretical_cdf = np.cumsum(p)
    empirical_cdf = np.cumsum(empirical_pmf)
    
    # Plot
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # PMF
    axes[0].bar(categories - 0.2, p, width=0.4, label='Theoretical PMF')
    axes[0].bar(categories + 0.2, empirical_pmf, width=0.4, alpha=0.7, label='Sample PMF')
    axes[0].set_xticks(categories)
    axes[0].set_title(f'PMF for probability set {i}')
    axes[0].legend()
    
    # CDF
    axes[1].step(categories, theoretical_cdf, where='mid', label='Theoretical CDF', color='blue')
    axes[1].step(categories, empirical_cdf, where='mid', label='Sample ECDF', color='orange')
    axes[1].set_xticks(categories)
    axes[1].set_title(f'CDF/ECDF for probability set {i}')
    axes[1].legend()
    
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Extract and clean data
x = nhanes['MaritalStatus'].dropna()

# Get empirical distribution
counts = x.value_counts(normalize=True)
categories = counts.index
empirical_pmf = counts.values

# Example theoretical probabilities (uniform)
p_theoretical = np.ones(len(categories)) / len(categories)
theoretical_cdf = np.cumsum(p_theoretical)
empirical_cdf = np.cumsum(empirical_pmf)

# Plot PMF / CDF
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# PMF comparison
axes[0].bar(np.arange(len(categories)) - 0.2, p_theoretical, width=0.4, label='Theoretical PMF')
axes[0].bar(np.arange(len(categories)) + 0.2, empirical_pmf, width=0.4, alpha=0.7, label='Empirical PMF')
axes[0].set_xticks(np.arange(len(categories)))
axes[0].set_xticklabels(categories, rotation=30, ha='right')
axes[0].set_ylabel('Probability')
axes[0].set_title('Marital Status PMF')
axes[0].legend()

# CDF comparison
axes[1].step(np.arange(len(categories)), theoretical_cdf, where='mid', label='Theoretical CDF', color='blue')
axes[1].step(np.arange(len(categories)), empirical_cdf, where='mid', label='Empirical ECDF', color='orange')
axes[1].set_xticks(np.arange(len(categories)))
axes[1].set_xticklabels(categories, rotation=30, ha='right')
axes[1].set_ylabel('Cumulative Probability')
axes[1].set_title('Marital Status CDF / ECDF')
axes[1].legend()

plt.tight_layout()
plt.show()
```
